```python
import glob
import os
import numpy as np
import sys
import random
from pathlib import Path
```

```python
data_dir = Path('/content/drive/Shareddrives/COS-470/Gestures/data/gestures_dataset')
valid_pct = 10;
test_pct = 20;
file_train = open('/content/drive/Shareddrives/COS-470/Gestures/data/gestures_dataset/train.txt', 'w')
file_train.truncate(None)
file_val = open('/content/drive/Shareddrives/COS-470/Gestures/data/gestures_dataset/val.txt', 'w')
file_val.truncate(None)
file_test = open('/content/drive/Shareddrives/COS-470/Gestures/data/gestures_dataset/test.txt', 'w')
file_test.truncate(None)

all_files = [str(path) for path in list(data_dir.glob('images/*.jpg'))]
valid_files = random.sample(all_files, k=round(len(all_files)*(valid_pct/100)))
all_files = [file for file in all_files if file not in valid_files]
test_files = random.sample(all_files, k=round(len(all_files)*(test_pct/100)))
all_files = [file for file in all_files if file not in test_files]

for path in valid_files:
  file_val.write(path + '\n')

for path in test_files:
  file_test.write(path + '\n')

for path in all_files:
  file_train.write(path + '\n')

file_train.close()
file_val.close()
file_test.close()
```
